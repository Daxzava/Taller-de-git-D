```python
import matplotlib.pyplot as plt
from numpy import *

def area(f, a, b):
  from sympy import symbols, integrate
  x, y = symbols('x y')
  A = integrate(f, (x, a, b))
  return A

a = 0                     # LIMITES DE INTEGRACIÓN
b = sqrt(2)

x = arange(-1, 2.1, 0.1)  # GRAFICACIÓN
y = x**3 - 2*x
plt.plot(x, y)
plt.fill_between(x, y, where = (x > a-0.1) & (x <= b))
plt.axhline(color='black')
plt.grid()
plt.show()

fx = 'x**3 -2*x'          # Función para
print('Area =', area(fx, a, b))
```

```python
import matplotlib.pyplot as plt
from numpy import *

def area(f, a, b):
  from sympy import symbols, integrate
  x, y = symbols('x y')
  A = integrate(f, (x, a, b))
  return A

a = -1
b = 1
c = 3                      # LIMITES DE INTEGRACIÓN
d = 7.1


x = arange(-2, 8.1, 0.1)  # GRAFICACIÓN
y = 0.02*x**4 - 0.2*x**3 + 0.4*x**2 + 0.2*x - 0.42
plt.plot(x, y)
plt.fill_between(x, y, where = (x > a-0.1) & (x <= b))
plt.fill_between(x, y, where = (x > c-0.1) & (x <= d))
plt.axhline(color='black')
plt.grid()
plt.show()

fx = '0.02*x**4 - 0.2*x**3 + 0.4*x**2 + 0.2*x - 0.42'          # Función para
print('Area1 =', area(fx, a, b))
print('Area2 =', area(fx, c, d))
```
